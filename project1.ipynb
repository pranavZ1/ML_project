```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('roberta_embeddings_english_abstractive.csv')
```

```python
df
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
Y = df.iloc[:,-1]
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X_scaled_df
```

```python
from collections import Counter
```

```python
print("Class Distribution:",Counter(Y))
```

### Class is balanced

```python
#PCA (Feature Selection)
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=300)
```

```python
X_pca = pca.fit_transform(X_scaled_df)
```

```python
print("PCA Completed! Explained Variance Ratio:", sum(pca.explained_variance_ratio_))
```

```python
processed_data = pd.DataFrame(X_pca)
```

```python
processed_data['Judgement Status'] = Y
```

```python
processed_data
```

```python
processed_data.to_csv('PCA_roberta_dataset.csv',index=False)
```
