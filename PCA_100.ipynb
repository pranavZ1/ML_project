```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('scaled_PCA_roberta.csv')
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.svm import LinearSVC
```

```python
from sklearn.pipeline import Pipeline
```

```python
X = df.iloc[:,:-1]
```

```python
pca = PCA(n_components=100,random_state=42)
```

```python
Y = df.iloc[:,-1]
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_processed = pd.DataFrame(X_pca)
```

```python
X_processed
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_processed)
```

```python
X_scaled_df = pd.DataFrame(X_scaled)
```

```python
X_scaled_df
```

```python
model_svc_linear = LinearSVC(C=0.1,max_iter=5000,random_state=42,dual=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled_df,Y,random_state=42,test_size=0.3)
```

```python
model_svc_linear.fit(X_train,Y_train)
```

```python
y_pred_svc_linear = model_svc_linear.predict(X_test)
```

```python
y_pred_svc_linear
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred_svc_linear,Y_test)*100,"%")
```
