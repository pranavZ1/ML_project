```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel('IJJD dataset.xlsx')
```

```python
df
```

```python
extractive_features = ['Extractive Summarized Judgements','Judgement Status']
```

```python
extractive_df = df[extractive_features]
```

```python
extractive_df
```

```python
from transformers import RobertaTokenizer, RobertaModel
```

```python
import torch
```

```python
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
```

```python
model = RobertaModel.from_pretrained("roberta-base")
```

```python
text_ext = extractive_df['Extractive Summarized Judgements'].to_list()
```

```python
text_ext
```

```python
tokens = tokenizer(
    text_ext,
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="pt"
)
```

```python
with torch.no_grad():
    outputs = model(
        input_ids=tokens["input_ids"],
        attention_mask=tokens["attention_mask"]
    )
```

```python
cls_embeddings = outputs.last_hidden_state[:, 0, :]
```

```python
attention_mask = tokens["attention_mask"]
expanded_mask = attention_mask.unsqueeze(-1).expand(outputs.last_hidden_state.size()).float()
sum_embeddings = torch.sum(outputs.last_hidden_state * expanded_mask, dim=1)
sum_mask = torch.clamp(attention_mask.sum(dim=1).unsqueeze(-1), min=1e-9)
mean_embeddings = sum_embeddings / sum_mask
```

```python
embeddings = cls_embeddings.numpy()
```

```python
labels = extractive_df['Judgement Status'].tolist() if 'Judgement Status' in extractive_df.columns else None
```

```python
embedding_df = pd.DataFrame(embeddings)
```

```python
if labels is not None:
    embedding_df['Judgement Status'] = labels  # Add labels if available
```

```python
embedding_df.to_csv('roberta_embeddings_english_extractive.csv', index=False)
```
