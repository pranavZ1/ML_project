```python
import pandas as pd
```

```python
data = pd.read_excel('IJJD dataset.xlsx')
```

```python
data
```

```python
data.columns
```

```python
texts = data['Abstractive Summarized Judgements (Kannada)'].tolist()
```

```python
labels = data['Judgement Status'].tolist() if 'Judgement Status' in data.columns else None
```

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
tokens = tokenizer(
    texts,
    padding = True,
    truncation = True,
    max_length=512,
    return_tensors='pt'
)
```

```python
from transformers import BertModel
```

```python
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
import torch
```

```python
with torch.no_grad():
    outputs = model(
        input_ids=tokens["input_ids"],
        attention_mask=tokens["attention_mask"]
    )
```

```python
cls_embeddings = outputs.last_hidden_state[:, 0, :]
```

```python
attention_mask = tokens["attention_mask"]
expanded_mask = attention_mask.unsqueeze(-1).expand(outputs.last_hidden_state.size()).float()
sum_embeddings = torch.sum(outputs.last_hidden_state * expanded_mask, dim=1)
sum_mask = torch.clamp(attention_mask.sum(dim=1).unsqueeze(-1), min=1e-9)
mean_embeddings = sum_embeddings / sum_mask
```

```python
import numpy as np
```

```python
embeddings = cls_embeddings.numpy()
```

```python
embedding_df = pd.DataFrame(embeddings)
```

```python
if labels is not None:
    embedding_df['Judgement Status'] = labels
```

```python
embedding_df.to_csv('bert_embeddings_kannada_abstractive.csv', index=False)
```
