```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('PCA_roberta_dataset.csv')
```

```python
df
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
Y = df.iloc[:,-1]
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X_scaled_df
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled_df,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# logistic regression
logi_model = LogisticRegression(solver = 'liblinear',max_iter=1000,random_state=42)
```

```python
logi_model.fit(X_train,Y_train)
```

```python
y_pred_logi = logi_model.predict(X_test)
```

```python
y_pred_logi
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy on test data using logistic regression")
accuracy_logi = accuracy_score(Y_test,y_pred_logi)*100
print(accuracy_logi,"%")
```

```python
y_pred_logi_train = logi_model.predict(X_train)
```

```python
print("Accuracy on training data using logistic regression")
print(accuracy_score(y_pred_logi_train,Y_train)*100,"%")
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train the LinearSVC model
svc_model = SVC(max_iter=5000, C=0.5,random_state=42,kernel = 'linear')  # Adjust C for regularization strength
svc_model.fit(X_train, Y_train)

# Evaluate the model
y_pred_svc = svc_model.predict(X_test)

print("LinearSVC:")
print("Accuracy:", accuracy_score(Y_test, y_pred_svc))
print("Classification Report:\n", classification_report(Y_test, y_pred_svc))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

# Create a pipeline with scaling and LinearSVC
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', LinearSVC(max_iter=5000, random_state=42))
])

# Train the model
pipeline.fit(X_train, Y_train)

# Evaluate the model
y_pred = pipeline.predict(X_test)
print(accuracy_score(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# Create a pipeline for scaling and SVM
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', LinearSVC(penalty='l2', max_iter=5000, random_state=42, dual=False))
])

# Use GridSearchCV to tune regularization parameter C
param_grid = {'svc__C': [0.01, 0.1, 1, 10, 100]}
grid = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, Y_train)

# Best parameters and performance
print(f"Best Parameters: {grid.best_params_}")
print(f"Training Accuracy: {grid.best_score_}")

# Test performance
y_pred = grid.best_estimator_.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred))
```
