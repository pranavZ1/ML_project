```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('roberta_embeddings_english_abstractive.csv')
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
Y = df.iloc[:,-1]
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
selector = SelectKBest(f_classif, k=200)  # Select top 100 features
X_train_selected = selector.fit_transform(X_train, Y_train)
X_test_selected = selector.transform(X_test)
```

```python
X_train_selected_df = pd.DataFrame(X_train_selected)
```

```python
X_train_selected_df
```

```python
from sklearn.svm import SVC
```

```python
from xgboost import XGBClassifier
```

```python
for i in [50,100,150,200]:
    pca = PCA(n_components=i)  # Retain 50 components
    X_train_reduced = pca.fit_transform(X_train_selected)
    X_test_reduced = pca.transform(X_test_selected)
    model = LogisticRegression(penalty='l2',max_iter=5000, random_state=42)
    model_svc = SVC(kernel='rbf')
    model_xgb = XGBClassifier(n_estimators=200, learning_rate=0.3, eval_metric='logloss', random_state=42)
    model.fit(X_train_reduced, Y_train)
    model_svc.fit(X_train_reduced, Y_train)
    model_xgb.fit(X_train_reduced, Y_train)
    print("Logistic Regression",i,model.score(X_test_reduced, Y_test))
    print("SVC",i,model_svc.score(X_test_reduced, Y_test))
    print("XGB",i,model_xgb.score(X_test_reduced, Y_test))
```

#### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Define the model
xgb = XGBClassifier(eval_metric='logloss')

# Define the parameter grid
param_grid = {
    'learning_rate': [0.15, 0.2, 0.25],
    'max_depth': [6, 7, 8],
    'n_estimators': [250, 300, 350],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.9, 1.0],
    'gamma': [0.15, 0.2, 0.25]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
grid_search.fit(X_train_reduced, Y_train)

# Get best parameters and accuracy
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier


# Apply PCA
pca = PCA(n_components=50)
X_train_pca = pca.fit_transform(X_train_selected)
X_test_pca = pca.transform(X_test_selected)

# Apply LDA
lda = LDA(n_components=min(len(set(Y_train)) - 1, X_train.shape[1]))  # n_classes - 1
X_train_lda = lda.fit_transform(X_train, Y_train)
X_test_lda = lda.transform(X_test)

# Train XGBoost on PCA-transformed data
xgb_pca = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_pca.fit(X_train_pca, Y_train)
y_pred_pca = xgb_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(Y_test, y_pred_pca)

# Train XGBoost on LDA-transformed data
xgb_lda = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_lda.fit(X_train_lda, Y_train)
y_pred_lda = xgb_lda.predict(X_test_lda)
accuracy_lda = accuracy_score(Y_test, y_pred_lda)

# Print results
print("Accuracy with PCA:", accuracy_pca)
print("Accuracy with LDA:", accuracy_lda)
```
