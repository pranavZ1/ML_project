## Deep Learning Model

```python
import tensorflow as tf
from keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,accuracy_score
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('scaled_PCA_roberta.csv')
```

```python
df
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
y_train_oh = tf.keras.utils.to_categorical(Y_train, num_classes=4)  # 4 classes
y_test_oh = tf.keras.utils.to_categorical(Y_test, num_classes=4)
```

```python
model = models.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dense(4, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train_oh, validation_split=0.1, epochs=20, batch_size=32, verbose=1)
```

```python
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)
```

```python
print("Classification Report:\n", classification_report(Y_test, y_pred_classes))
```

```python
print("Accuracy On Deep-Learning Model : ",accuracy_score(Y_test,y_pred_classes)*100,"%")
```
