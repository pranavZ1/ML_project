```python
abstractive_features = ['Extractive Summarized Judgements (Tamil)','Judgement Status']
```

```python
abstractive_df = df[abstractive_features]
```

```python
abstractive_df
```

```python
from transformers import RobertaTokenizer, RobertaModel
```

```python
import torch
```

```python
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
```

```python
text_abs = abstractive_df['Extractive Summarized Judgements (Tamil)'].to_list()
```

```python
text_abs
```

```python
tokens = tokenizer(
    text_abs,
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="pt"
)
```

```python
model = RobertaModel.from_pretrained("roberta-base")
```

```python
with torch.no_grad():
    outputs = model(
        input_ids=tokens["input_ids"],
        attention_mask=tokens["attention_mask"]
    )
```

```python
cls_embeddings = outputs.last_hidden_state[:, 0, :]
```

```python
attention_mask = tokens["attention_mask"]
expanded_mask = attention_mask.unsqueeze(-1).expand(outputs.last_hidden_state.size()).float()
sum_embeddings = torch.sum(outputs.last_hidden_state * expanded_mask, dim=1)
sum_mask = torch.clamp(attention_mask.sum(dim=1).unsqueeze(-1), min=1e-9)
mean_embeddings = sum_embeddings / sum_mask
```

```python
import numpy as np
```

```python
embeddings = cls_embeddings.numpy()
```

```python
labels = abstractive_df['Judgement Status'].tolist() if 'Judgement Status' in abstractive_df.columns else None
```

```python
embedding_df = pd.DataFrame(embeddings)
```

```python
if labels is not None:
    embedding_df['Judgement Status'] = labels  # Add labels if available
```

```python
embedding_df.to_csv('roberta_embeddings_tamil_extractive.csv', index=False)
```
